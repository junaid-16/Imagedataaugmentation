```
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
```

function used for showing augmented image

```
def show_images(generator):
    generator.fit(images)
    image_iterator = generator.flow(images)

    fig, rows = plt.subplots(nrows=1, ncols=4, figsize=(18,18))
    for row in rows:
        row.imshow(image_iterator.next()[0].astype('int'))
        row.axis('off')
    plt.show()
```

loading image

```
image = plt.imread("cat.jpg")
images = image.reshape((1, image.shape[0], image.shape[1], image.shape[2]))

plt.imshow(images[0])
plt.show()
```

```
generator=ImageDataGenerator(rotation_range=40)
show_images(generator)
```

```
generator=ImageDataGenerator(width_shift_range=0.3)
show_images(generator)
```

```
generator=ImageDataGenerator(width_shift_range=0.3,fill_mode='reflect')
show_images(generator)
```

```
generator=ImageDataGenerator(height_shift_range=0.2,fill_mode='constant')
show_images(generator)
```

```
generator=ImageDataGenerator(width_shift_range=0.3,fill_mode='wrap')
show_images(generator)
```

```
generator=ImageDataGenerator(brightness_range=(0.3,0.8))
show_images(generator)
```

```
generator=ImageDataGenerator(shear_range= 40)
show_images(generator)
```

```
generator=ImageDataGenerator(zoom_range=0.5)
#generator=ImageDataGenerator(zoom_range=(0.2,0.5))
show_images(generator)
```

```
generator=ImageDataGenerator(channel_shift_range=90)
show_images(generator)
```

```
generator=ImageDataGenerator(horizontal_flip=True)
show_images(generator)
```

```
generator=ImageDataGenerator(vertical_flip=True)
show_images(generator)
```
